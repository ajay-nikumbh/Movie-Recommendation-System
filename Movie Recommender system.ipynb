#### 9. Helper method for doing that task

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies
```

#### 10. Helper function to retrive the top 3 actors names

```python
def convert3(obj):
    L=[]
    counter=0
    
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter += 1
        else:
            break
    
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies
```

```python
movies['crew'][0]
```

#### 11. Helper function to fetch director names

```python
def fetch_director(obj):
    l=[]
    
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
            
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

#### 4. Nlp operations

#### 1. Apply stemming on the tags feature

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps= PorterStemmer()
ps
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

#### 2. Count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.size
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

#### 5. Distance computation - cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity =cosine_similarity(vectors)
```

```python
len(similarity[0])
```

#### 5. Make the recommend function

```python
def recommend(movie):
    
    # Fetch the index
    movie_index= new_df[new_df['title'] == movie].index[0]
    
    # Fetch the distances
    distances = similarity[movie_index]
    
    # Get the 5 similar movies 
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    # Now print the first indexe's of the top 5 similar movies
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        

        
```

```python
recommend("Avatar")
```
